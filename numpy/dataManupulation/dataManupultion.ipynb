```python
#data manupultion using pandas 
#sample data 
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, None, 35, 40, 45],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'life_expectancy': [None, 75, 78, 82, 77],
    'fun_score': [7, 8, 6, 9, 5],
    'climate_score': [6, 7, None, 8, 4]
})

print(df)
```

```python
#fining misiing values
print(df.isnull())

#filling missing values
df_filled = df.fillna(0)
print(df_filled)
```

```python
#filling missing values with mean of the column
df_filled_mean = df.fillna(df.mean(numeric_only=True))
print(df_filled_mean)
```

```python
#renaming colomns
df_renamed = df.rename(columns={'Name': 'Full Name', 'Age': 'Age in Years'})
print(df_renamed)
```

```python
#data aggregating and grouping
#grouping by city and calculating mean age
grouped = df.groupby('City')['Age'].mean()
print(grouped)

grouped_fun_score = df.groupby('City')['fun_score'].mean()
print(grouped_fun_score)

#grouping by mullltiple columns and calculating mean
grouped_multi = df.groupby(['City', 'climate_score'])['fun_score'].mean()
print(grouped_multi)
```

```python
#mergign and joining dataframes
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})
df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Age': [25, 30, 35]
})
#merging on ID column using inner join
merged_inner = pd.merge(df1, df2, on='ID', how='inner')
print(merged_inner)

#similarly we can do left, right and outer join using how parameter in merge function.
```
