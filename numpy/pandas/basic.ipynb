```python
#pandas 
# it is a library in python which is used for data manipulation and analysis. It provides data structures like Series and DataFrame, which are built on top of numpy arrays, to handle and analyze structured data efficiently. 
#series is a one-dimensional array-like object that can hold any data type, while DataFrame is a two-dimensional table-like data structure that can hold multiple series with different data types.
#dataFrame is a 2D data structure that can hold multiple series with different data types. It is similar to a spreadsheet or SQL table, and it provides powerful tools for data manipulation, cleaning, and analysis.
```

```python
import pandas as pd

#series from list 
arr = pd.Series([1,2,3,4,5])
print(arr)

#series from dict
data = {'a': 1, 'b': 2, 'c': 3}
arr2 = pd.Series(data)
print(arr2)

#here we create a series from an list and a dict , list got its custom index and dict got its keys as index.

#we can also give custom index to series created from list and dict
arr3 = pd.Series([1,2,3,4,5], index=['a', 'b', 'c', 'd', 'e'])
print(arr3) 
```

```python
#dataFrame from dict of list
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'City': ['New York', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)
print(df) 

#similarly we can create dataFrame from dict of series, list of dict, list of list, dict of dict etc.
```

```python
#accessing data from dataFrame
print(df['Name']) #accessing a column
print("------------------")
print(df[['Name', 'Age']]) #accessing multiple columns
print("------------------")
print(df.loc[0]) #accessing a row by index
print("------------------")
print(df.iloc[0]) #accessing a row by position
print("------------------")
print(df.loc[0, 'Name']) #accessing a specific value by row index and column name
print("------------------")
print(df.iloc[0, 0]) #accessing a specific value by row position and column position
print("------------------")
#accessing by specific element 
print(df.at[0, 'Name']) #accessing a specific value by row index and column name
print("------------------")
print(df.iat[0, 0]) #accessing a specific value by row position and column position
```

```python
#data manipulation in dataFrame
#adding a new column
df['Country'] = ['USA', 'USA', 'USA']
print(df)
print("------------------")
#deleting a column
df = df.drop(['Country'], axis=1) #axis 1 is for columns, axis 0 is for rows
print(df)
print("------------------")


```
