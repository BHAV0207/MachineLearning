```python
##Python is a dynamically typed language, which means that you don't have to declare the type of a variable when you create it. The type of a variable is determined at runtime based on the value it is assigned. This allows for more flexibility in your code, but it also means that you need to be careful when working with variables, as they can change type unexpectedly.

##For example, you can assign an integer value to a variable and then later assign a string value to the same variable:
x = 10  # x is an integer
print(x)  # Output: 10
x = "Hello, World!"  # x is now a string
print(x)  # Output: Hello, World!
```

```python
## For input we use the input() function. It takes a string as an argument, which is displayed to the user as a prompt. The function then waits for the user to enter some input and returns it as a string.
name = input("What is your name? ")
print("Hello, " + name + "!")

## it always return a string, even if the user enters a number. If you want to convert the input to a different type, you can use the appropriate type conversion function, such as int() for integers or float() for floating-point numbers.
age = input("How old are you? ")
age = int(age)  # Convert the input to an integer
print("You are " + str(age) + " years old.")  
```
