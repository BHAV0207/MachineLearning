```python
#instance variables and insrance methods
class Dog:
  def __init__(self, name , age): # its not necessary to have the self parameter but its a convention to use it , we can use any name instead of self but we have to use the same name in the rest of the code
    self.name = name
    self.age = age #instance variables are variables that are specific to an instance of a class , they are defined in the __init__ method and are accessed using the self parameter
  
  def bark(self): #instance method is a function that is defined in a class and is called on an instance of the class , it can access the instance variables and can perform operations on them
    print( f"{self.name} says Woof!")

labra = Dog("bruno", 5)
print(labra.name) #bruno
print(labra.age) #5
labra.bark() #bruno says Woof!
```

```python
#inheritance 
class Car:
  def __init__(self , doors , windows , type):
    self.doors = doors
    self.windows = windows
    self.type = type
  def start(self):
    print("Car is starting...")
    
class Tesla(Car):
  def __init__(self , doors , windows , type , self_driving):
    super().__init__(doors , windows , type) #super() is used to call the __init__ method of the parent class
    self.self_driving = self_driving #self_driving is an instance variable that is specific to the Tesla class
    
  def start(self):
      print("Tesla is starting...") #overriding the start method of the Car class
    
  def autopilot(self):
      print("Tesla is in autopilot mode...") #new method that is specific to the Tesla class
      
tesla_model_s = Tesla(4 , 4 , "sedan" , True)
print(tesla_model_s.doors) #4
print(tesla_model_s.windows) #4
print(tesla_model_s.type) #sedan
print(tesla_model_s.self_driving) #True
tesla_model_s.start() #Tesla is starting...
tesla_model_s.autopilot() #Tesla is in autopilot mode...  
```

```python
#multiple inheritance
class Electric:
  def __init__(self , battery_capacity):
    self.battery_capacity = battery_capacity
    
  def charge(self):
    print("Electric car is charging...")

class Hybrid:
  def __init__(self , fuel_capacity):
    self.fuel_capacity = fuel_capacity
    
  def refuel(self):
    print("Hybrid car is refueling...")

class TeslaModelX(Tesla , Electric , Hybrid):
  def __init__(self , doors , windows , type , self_driving , battery_capacity , fuel_capacity):
    Tesla.__init__(self , doors , windows , type , self_driving) #calling the __init__ method of the Tesla class
    Electric.__init__(self , battery_capacity) #calling the __init__ method of the Electric class
    Hybrid.__init__(self , fuel_capacity) #calling the __init__ method of the Hybrid class
  
tesla_model_x = TeslaModelX(4 , 4 , "SUV" , True , 100 , 50)
print(tesla_model_x.doors) #4
print(tesla_model_x.windows) #4
print(tesla_model_x.type) #SUV
print(tesla_model_x.self_driving) #True
print(tesla_model_x.battery_capacity) #100
print(tesla_model_x.fuel_capacity) #50
tesla_model_x.start() #Tesla is starting...
tesla_model_x.charge() #Electric car is charging...
tesla_model_x.refuel() #Hybrid car is refueling...
```

```python
#polymorphism
#polymorphism is the ability of an object to take on many forms. In Python, polymorphism is achieved through method overriding and duck typing. Duck typing is a concept where the type of an object is determined by its behavior rather than its class. This means that if an object can perform a certain action, it can be treated as that type of object regardless of its actual class.


#methos overriding is a form of polymorphism where a child class can provide a specific implementation of a method that is already defined in its parent class
class Animal:
  def speak(self):
    print("Animal is speaking...")
    
class Cat(Animal):
  def speak(self):
    print("Cat says Meow!")
    
#here we have a parent class Animal and a child class Cat that inherits from the Animal class. The Cat class overrides the speak method of the Animal class to provide a specific implementation for cats.
animal = Animal()
cat = Cat()
animal.speak() #Animal is speaking...
cat.speak() #Cat says Meow!
```

```python
#polymorphism with functions and methods

class Shape:
  def area(self):
    return "area of the figure..."
  
class Circle(Shape):
  def __init__(self , radius):
    self.radius = radius
    
  def area(self):
    return 3.14 * self.radius * self.radius

class Rectangle(Shape):
  def __init__(self , length , width):
    self.length = length
    self.width = width
    
  def area(self):
    return self.length * self.width

#functions that demonstrate polymorphism
def print_area(shape):
  print(shape.area())

shapes = [Circle(5) , Rectangle(4 , 6)]
for shape in shapes:
  print_area(shape) #78.5 , 24
```

```python
#abstract classes and interfaces
#abstract classes are classes that cannot be instantiated and are meant to be subclassed. They can contain abstract methods, which are methods that are declared but not implemented in the abstract class. Subclasses of the abstract class must implement the abstract methods. In Python, we can use the abc module to create abstract classes and define abstract methods.
from abc import ABC , abstractmethod
class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Circle(Shape):
  def __init__(self , radius):
    self.radius = radius
    
  def area(self):
    return 3.14 * self.radius * self.radius

class Rectangle(Shape):
  def __init__(self , length , width):
    self.length = length
    self.width = width
    
  def area(self):
    return self.length * self.width 
  
circle = Circle(5)
rectangle = Rectangle(4 , 6)
print(circle.area()) #78.5  
print(rectangle.area()) #24
```

```python
#Encaptulation 
#Encapsulation is the process of hiding the internal details of an object and only exposing a public interface. This is done to protect the data and to prevent unauthorized access to it. In Python, we can achieve encapsulation by using private variables and methods. Private variables and methods are defined with a double underscore prefix (__) and are not accessible from outside the class.

class Person:
  def __init__(self , name , age):
    self.__name = name #private variable
    self.__age = age #private variable
    
  def get_name(self): #public method to access the private variable __name
    return self.__name
  
  def get_age(self): #public method to access the private variable __age
    return self.__age

person = Person("Alice" , 30)
print(person.get_name()) #Alice
print(person.get_age()) #30



#protected variables and methods
#Protected variables and methods are defined with a single underscore prefix (_) and are meant to be accessed from within the class and its subclasses. They are not meant to be accessed from outside the class, but they can be accessed if needed.

class Employee:
  def __init__(self , name , salary):
    self._name = name #protected variable
    self._salary = salary #protected variable
    
  def get_name(self): #public method to access the protected variable _name
    return self._name
  
  def get_salary(self): #public method to access the protected variable _salary
    return self._salary
  
class Manager(Employee):
  def __init__(self , name , salary , department):
    super().__init__(name , salary) #calling the __init__ method of the Employee class
    self.department = department #new variable that is specific to the Manager class
  

manager = Manager("Bob" , 50000 , "Sales")
print(manager.get_name()) #Bob
print(manager.get_salary()) #50000
print(manager.department) #Sales
```

```python
#Abstraction
#Abstraction is the process of hiding the implementation details of an object and only exposing the essential features. This is done to reduce complexity and to make the code easier to understand. In Python, we can achieve abstraction by using abstract classes and interfaces. Abstract classes are classes that cannot be instantiated and are meant to be subclassed. They can contain abstract methods, which are methods that are declared but not implemented in the abstract class. Subclasses of the abstract class must implement the abstract methods. In Python, we can use the abc module to create abstract classes and define abstract methods.
from abc import ABC , abstractmethod
class Shape(ABC):
  @abstractmethod
  def area(self):
    pass
  
  
class Circle(Shape):
  def __init__(self , radius):
    self.radius = radius
    
  def area(self):
    return 3.14 * self.radius * self.radius
  
  
class Rectangle(Shape):
  def __init__(self , length , width):
    self.length = length
    self.width = width
    
  def area(self):
    return self.length * self.width

circle = Circle(5)
rectangle = Rectangle(4 , 6)
print(circle.area()) #78.5
print(rectangle.area()) #24
```
