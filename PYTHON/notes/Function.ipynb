```python
#syntax of the function 
def function_name(parameters):
    # function body
    return value
  
#default parameter value
def greet(name="World"):
    return f"Hello, {name}!"  
#calling the functionprint(greet())  # Output: Hello, World!
print(greet("Alice"))  # Output: Hello, Alice!
print(greet())  


#positional arguments
def add(a, b):
    return a + b  
print(add(5, 3))  # Output: 8

#keyword arguments
def introduce(**kwargs):
  for key, value in kwargs.items():
        print(f"{key}: {value}")
introduce(name="Alice", age=30, city="New York")

#variable-length arguments
def sum_all(*args):
    return sum(args)
print(sum_all(1, 2, 3, 4))  # Output: 10


#return multiple values
def get_coordinates():
    x = 10
    y = 20
    return x, y
coordinates = get_coordinates()
print(coordinates)  # Output: (10, 20)
```

```python
#question 1 : converting temperature

def temp_converter(temp , unit):
  if unit =='C' or unit == 'c' :
    return temp *9/5 +32
  else:
    return (temp - 32) * 5/9
  
temp = float(input("Enter the temperature: "))
print(temp_converter(temp, 'C'))  # Convert to Fahrenheit
print(temp_converter(temp, 'F'))  # Convert to Celsius
```

```python
#question 2 : password strength checker 
def pass_checker(pass):
  if len(pass) < 8:
    return False
  if not any(char.isdigit() for char in pass):
    return False
  if not any(char.isupper() for char in pass):
    return False
  if not any(char.islower() for char in pass):
    return False
  return True
```

```python
#checking string is a palindrome or not
def is_palindrome(str):
  str = str.replace(" ", "").lower()
  return str == str[::-1]
print(is_palindrome("A man a plan a canal Panama"))  # Output: True
print(is_palindrome("Hello"))  # Output: False
```
