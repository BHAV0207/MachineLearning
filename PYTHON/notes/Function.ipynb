````python
#question 1 : converting temperature

def temp_converter(temp , unit):
  if unit =='C' or unit == 'c' :
    return temp *9/5 +32
  else:
    return (temp - 32) * 5/9
  
temp = float(input("Enter the temperature: "))
print(temp_converter(temp, 'C'))  # Convert to Fahrenheit
print(temp_converter(temp, 'F'))  # Convert to Celsius
````

````python
#question 2 : password strength checker 
def pass_checker(pass):
  if len(pass) < 8:
    return False
  if not any(char.isdigit() for char in pass):
    return False
  if not any(char.isupper() for char in pass):
    return False
  if not any(char.islower() for char in pass):
    return False
  return True
````

````python
#checking string is a palindrome or not
def is_palindrome(str):
  str = str.replace(" ", "").lower()
  return str == str[::-1]
print(is_palindrome("A man a plan a canal Panama"))  # Output: True
print(is_palindrome("Hello"))  # Output: False
````

``` Lambda Functions ```

````python
#these functions are defined using the lambda keyword, which allows you to create small, anonymous functions in a single line of code.
#there can be any number of arguments but only one expression, which is evaluated and returned.
lambda arguments: expression

#lambda function to add two numbers
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8

#lambda function to find the square of a number
square = lambda x: x ** 2
print(square(4))  # Output: 16

#square of all teh numbers in a list using lambda function
numbers = [1, 2, 3, 4, 5]
list(map(lambda x:x**2,numbers))
````

````python
#lambda function WITH MAP FUNCTION
#we can map multiple iterators using map function
numbers1 = [1, 2, 3, 4, 5]
numbers2 = [6, 7, 8, 9, 10]
result = list(map(lambda x, y: x + y, numbers1, numbers2))
print(result)  # Output: [7, 9, 11, 13, 15]
````

````python
#map function syntax 

#map(function, iterable)

#you can use the map function to apply a function to each item in an iterable (like a list) and return a new iterable with the results.
#the function can be a built-in function, a user-defined function, or a lambda function. The map function can also take multiple iterables as arguments, and the function will be applied to the corresponding items from each iterable.

#fileter function syntax

#filter(function, iterable)

#the filter function is used to filter items from an iterable based on a specified condition. It takes a function and an iterable as arguments and returns an iterator that contains only the items from the iterable for which the function returns True.
#the function can be a built-in function, a user-defined function, or a lambda function. The filter function is often used to create a new list that contains only the items that meet a certain condition.

#filter function to filter out even numbers from a list
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4, 6]

#filter function with lambda function and multiple conditions
numbers = [1, 2, 3, 4, 5, 6]
filtered_numbers = list(filter(lambda x: x % 2 == 0 and x > 3, numbers))
print(filtered_numbers)  # Output: [4, 6]


````
