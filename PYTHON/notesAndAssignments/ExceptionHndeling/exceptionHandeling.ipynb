```python
#exception handeling is a way to handle errors in python. It allows you to write code that can handle errors gracefully without crashing the program. The try block is used to wrap the code that may raise an exception, and the except block is used to handle the exception if it occurs. The else block is executed if no exception is raised, and the finally block is executed regardless of whether an exception is raised or not.

#exception handeling in python
try:
    x = 10 / 0 #this will raise a ZeroDivisionError
except ZeroDivisionError:
    print("You cannot divide by zero") #this will be executed if a ZeroDivisionError is raised
except Exception as e:
    print("An error occurred:", e) #this will be executed if any other exception is raised
else:
    print("The division was successful") #this will be executed if no exception is raised
finally:
    print("This will always be executed") #this will be executed regardless of whether an exception is raised or not
    
```
