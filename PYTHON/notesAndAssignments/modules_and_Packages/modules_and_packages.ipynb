```python
#some importanrt modules in python
#1. math module - to perform mathematical operations
#2. random module - to generate random numbers
import random
print(random.randint(1, 10)) #to generate a random number between 1 and 10

#3. datetime module - to work with dates and times
#4. os module - to work with operating system
import os
print(os.getcwd()) #to get the current working directory  


#5. json module - to work with json data
import json
data = {'name': 'John', 'age': 30, 'city': 'New York'}
json_data = json.dumps(data) #to convert python object to json string
print(json_data)
python_data = json.loads(json_data) #to convert json string to python object
print(python_data)

#6. re module - to work with regular expressions
import re
text = "The rain in Spain stays mainly in the plain"
x = re.findall("ai", text) #to find all the occurrences of "ai"
print(x)
```
