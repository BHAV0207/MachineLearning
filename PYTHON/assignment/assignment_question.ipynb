**Question 4:** Write a Python program to find the largest of three numbers.

```python
# Your code here
input_num_1 =int(input("Enter a number: "))
input_num_2 =int(input("Enter another number: "))
input_num_3 =int(input("Enter a third number: "))
if input_num_1 > input_num_2 and input_num_1 > input_num_3:
    print("The largest number is:", input_num_1)
elif input_num_2 > input_num_1 and input_num_2 > input_num_3:
    print("The largest number is:", input_num_2)
else:
    print("The largest number is:", input_num_3)
```

**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here
input_num =int(input("Enter a number: "))
for i in range(input_num-1, 1, -1):
    input_num *= i
print("The factorial is:", input_num)
```
